```python
# DESCRIPTION: Aggregates constructor performance for the 'Constructor Wars' dashboard
# ---------------------------------------------------------

from pyspark.sql.functions import sum as _sum, col

# 1. Read Silver Data
df = spark.read.format("delta").load("abfss://silver@YOUR_ACCOUNT.dfs.core.windows.net/f1_results")

# 2. Aggregate Points by Team & Season
constructor_dominance_df = df.groupBy("season", "team_name") \
    .agg(_sum("actual_points").alias("total_points"), _sum("position").alias("total_position_score")) \
    .orderBy(col("season").desc(), col("total_points").desc())

# 3. Save to Gold
constructor_dominance_df.write.mode("overwrite").format("parquet").save("abfss://gold@YOUR_ACCOUNT.dfs.core.windows.net/constructor_dominance")
```
